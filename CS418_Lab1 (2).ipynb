**Exercise 1.1** Write a python function that returns the sum of squares in range  [1, 2, ..., n ]

```python
def sqrsum(a):
    b=0
    for i in range(1,a+1):
        b = b + (i * i);
    return b
a=5
print("Sum of squares is=")
print(sqrsum(a))
```

**Exercise 1.2** Write a lambda function that returns the sum of squares in range  [1, 2, ..., n ]

```python
numbers = []
def sqrsum(a):
    for i in range(1,a+1):
        numbers.append(i)
    return numbers
a=5

y = sqrsum(a)
x = list(map(lambda n: n ** 2, y))

sum = 0
for ele in x:
    sum += ele
print("Sum of squares using lambda function=")
print(sum)
```

**Exercise 2.1** For a given string s, write a function that returns a new string that doubles
each character of the original string. For instance, if the user enters _Python_, the output should be
_PPyytthhoonn_.

```python
def double_letters (s):
    a = ''
    for character in s:
        a += character + character
    return a
    
s = 'Hello'
print("Expected string is=")
print(double_letters(s))    
```

```python
# 1.Turn the whole string to lowercase
with open('Yellow_Wallpaper_PROJECT GUTENBERG.txt', 'r') as fileinput:
    for line in fileinput:
        line = line.lower()
        print(line)
```

```python
# 2.Remove all the punctuations from the text (you can use "from string import punctuation")
with open('Yellow_Wallpaper_PROJECT GUTENBERG.txt', 'r') as fileinput:

    p ='''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    final_str = ''

    for line in fileinput:
        for letter in line:
            if letter in p:
                abc = letter.replace(letter,"")
            else:
                abc = letter
            final_str += abc
    print(final_str)
```

```python
# 3.Split the string into a list of individual words
with open('Yellow_Wallpaper_PROJECT GUTENBERG.txt', 'r') as fileinput:


    for line in fileinput:
        words=line.split()
        for word in words:
            print(word)
    
    
```

```python
# 4. Create a dictionary and count the frequency of words
with open('Yellow_Wallpaper_PROJECT GUTENBERG.txt', 'r') as fileinput:
    wordfreq={}
    p = ['.',',','!','-']
    final_str = ''
    for line in fileinput:
        line = line.lower()
        words=line.split()
        for word in words:
                final_str = ''
                for letter in word:
                    if letter in p:
                        abc = letter.replace(letter,"")
                    else:
                        abc = letter
                    final_str += abc
                if final_str not in wordfreq:
                    wordfreq[final_str]=1
                else:
                    wordfreq[final_str] += 1
            
    print(wordfreq)
```
